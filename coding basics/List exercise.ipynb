Let us say your expense for every month are listed below,
January - 2200
February - 2350
March - 2600
April - 2130
May - 2190
Create a list to store these monthly expenses and using that find out,

1. In Feb, how many dollars you spent extra compare to January?
2. Find out your total expense in first quarter (first three months) of the year.
3. Find out if you spent exactly 2000 dollars in any month
4. June month just finished and your expense is 1980 dollar. Add this item to our monthly expense list
5. You returned an item that you bought in a month of April and
got a refund of 200$. Make a correction to your monthly expense list

```python
monthly_expense= [{"mon":"JAN", "expense":2200},
                  {"mon":"FEB", "expense":2350},
                  {"mon":"MAR", "expense":2600},
                  {"mon":"APR", "expense":2130},
                  {"mon":"MAY", "expense":2190}]
monthly_expense
```

```python
print(f"Monthly extra expense in Feb: {monthly_expense[1].get("expense") - monthly_expense[0].get("expense")}")
```

```python
print(f"Monthly extra expense in Feb: {monthly_expense[2].get("expense")+ monthly_expense[1].get("expense") + monthly_expense[0].get("expense")}")
```

```python
for i in range(len(monthly_expense)):
    if monthly_expense[i].get("expense") == 2000:
        print(monthly_expense[i].get("mon"))
    else: 
        print("Not present")
```

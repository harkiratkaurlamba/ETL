```python
a = [1,2,3,4,2,2,5,1]
a
```

```python
d ={}
for i in range(len(a)):
    if a[i] in d:
        d[a[i]] = d[a[i]]+1
    else:
        d[a[i]]=1
    #print(f"i is {i}, a[{i}] is: {a[i]} and d is: {d}")
d
```

```python
l = ['apple','bat','bar','book','atom']
d={}
for word in l:
    if word[0] in d:
        d[word[0]].append(word)
    else:
        d[word[0]]=[word]
d
```

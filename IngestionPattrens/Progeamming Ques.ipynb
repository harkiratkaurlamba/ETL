Method 1: Using set() (most common)

```python
nums = [4, 1, 2, 2, 5, 5, 3]

unique_nums = list(set(nums))
unique_nums.sort(reverse=True)

second_highest = unique_nums[1]
print(second_highest)
```

Method 2: Single pass (no sorting – better for large data)

```python
nums = [4, 1, 2, 2, 5, 5, 3]

highest = second = float('-inf')

for n in nums:
    if n > highest:
        second = highest
        highest = n
    elif highest > n > second:
        second = n

print(second)
```

Method 3: Pythonic (interview friendly)

```python
nums = [4, 1, 2, 2, 5, 5, 3]
print(sorted(set(nums))[-2])
```

```python
%sql
SELECT MAX(value) AS second_highest
FROM numbers
WHERE value < (SELECT MAX(value) FROM numbers);
```

```python
%sql
SELECT value
FROM (
    SELECT value, DENSE_RANK() OVER (ORDER BY value DESC) AS rnk
    FROM numbers
) t
WHERE rnk = 2;
```

Edge Case Handling 
- If all numbers are same → no 2nd highest
- If only one unique value exists → return NULL / handle exception

```python
employees = [
    (1, "Alice"),
    (2, "Bob"),
    (3, "Charlie"),
    (4, "David")
]

salaries = [
    (1, 70000),
    (2, None),
    (3, 90000),
    (4, 90000)
]

emp_df = spark.createDataFrame(employees, ["emp_id", "name"])
sal_df = spark.createDataFrame(salaries, ["emp_id", "salary"])

emp_df.show()
sal_df.show()

joined_df = (
    emp_df.join(sal_df, on="emp_id", how="inner")
          .filter(col("salary").isNotNull())
)

joined_df.show()


window_spec = Window.orderBy(col("salary").desc())

highest_paid_df = (
    joined_df
    .withColumn("rank", dense_rank().over(window_spec))
    .filter(col("rank") == 1)
    .select("emp_id", "name", "salary")
)

highest_paid_df.show()

```

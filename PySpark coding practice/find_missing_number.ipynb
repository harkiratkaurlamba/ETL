```python
def find_missing_number(nums, n):
    """
    Find the missing number in the list assuming sequence from 1 to n.
    
    Args:
    nums (list): List of integers with one missing number.
    n (int): Expected length of complete sequence.
    
    Returns:
    int: The missing number.
    """
    expected_sum = n * (n + 1) // 2  # Gauss formula [web:1]
    actual_sum = sum(nums)
    return expected_sum - actual_sum

# Example usage
nums = [1, 2, 4, 5, 6]
n = 6
missing = find_missing_number(nums, n)
print(f"Missing number: {missing}")  # Output: 3
```

Batch schema evolution (Delta)

Enable schema evolution (append new columns)

```python
df.write \
  .format("delta") \
  .mode("append") \
  .option("mergeSchema", "true") \
  .save("/delta/events")
```

Overwrite schema (replace existing)

```python
df.write \
  .format("delta") \
  .mode("overwrite") \
  .option("overwriteSchema", "true") \
  .save("/delta/events")
```

Streaming schema evolution (Delta)

Append with schema evolution

```python
spark.conf.set("spark.databricks.delta.schema.autoMerge.enabled", "true")

df_stream.writeStream \
  .format("delta") \
  .outputMode("append") \
  .option("checkpointLocation", "/chk/events") \
  .start("/delta/events")
```

Key rules (important)
- Only additive changes supported automatically (new columns)
- Type changes / column drops ❌ not automatic
- Streaming requires autoMerge enabled
- Prefer evolution at Bronze → Silver boundary

One-liner:
Batch uses mergeSchema, streaming needs autoMerge, both support only additive schema changes.

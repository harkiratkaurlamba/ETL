Manual Loop (Educational)
This method is straightforward but slower for large lists due to repeated in checks

```python
original_list = [1, 2, 2, 3, 4, 4, 5]
unique_list = []
for item in original_list:
    if item not in unique_list:
        unique_list.append(item)
print(unique_list) 
```

Fastest (No Order)
Convert to a set() for optimal speed on large lists, though order isn't guaranteed

```python
original_list = [1, 2, 2, 3, 4, 4, 5]
unique_list = list(set(original_list))
print(unique_list)  # Output: [1, 2, 3, 4, 5] (order may vary)
```

Preserve Order (Recommended)
Use dict.fromkeys() to maintain the original sequence while eliminating repeats, as dictionaries (since Python 3.7) preserve insertion order.

This approach is concise, readable, and performs well for most cases without requiring additional imports.


```python
original_list = [1, 2, 2, 3, 4, 4, 5]
unique_list = list(dict.fromkeys(original_list))
print(unique_list)  # Output: [1, 2, 3, 4, 5]
```

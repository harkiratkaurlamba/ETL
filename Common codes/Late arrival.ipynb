```python
from pyspark.sql.types import DateType
dataset=[('Ava', '2026-03-02'),
         ('Bob', '2025-03-02'),
         ('Charles', '2026-05-31'),
         ('Dentu', '2024-03-02'),
         ('Emma', '2026-01-07')]
col=['name' ,'datetime']

df = spark.createDataFrame(dataset, col)
df.withColumn("datetime", df.datetime.cast("DateType()"))
display(df)
df.printSchema()
```

```python
from datetime import date

cutoff_date = date(2026,1,1)

df2=df.selectExpr("name", "cast(datetime as date)")
df3 = df2.where(f"datetime >= '{cutoff_date}'")
```

```python
df3.show()
```

```python
df3.write.format('delta').mode('append').save('/Volumes/hk/playground/messwithme/LateArrivalData')
```

```python
from pyspark.sql import SparkSession
import datetime
from dbxscd import SCD3  # Assuming library imported as dbxscd

spark = SparkSession.builder.getOrCreate()
df = spark.createDataFrame([
    ('FA', 'Fiji Apple', 'Red', 3.5),
    ('BN', 'Banana', 'Yellow', 1.0)
], ['ShortName', 'Fruit', 'Color', 'Price'])

current_time = datetime.datetime.now().strftime("%Y-%m-%dT%H:%M:%S")
target_table = 'default.customer_scd3'
partition_keys = ['ShortName']
key_cols = 'ShortName, Fruit'

SCD3(spark, df, target_table, partition_keys, key_cols, current_time)
spark.sql(f"SELECT * FROM {target_table}").display()
```
